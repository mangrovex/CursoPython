# Ejes y mapeados

En esta segunda lección aprenderemos a:

* Mapear texto en los ejes numéricos
* A trabajar con el eje X y el eje Y
* A crear gráficos con los ejes invertidos

```python
import matplotlib.pyplot as plt
%matplotlib inline
```



## Mapeado de texto en los ejes

Por defecto en los gráficos no podemos mostrar texto en uno de los ejes, ya que éstos se basan en valores numéricos. 

Lo que se hace es mapear los textos a mostrar en una lista del mismo tamaño que el número de valores de un eje (x para el horizontal o y para el vertical) y con el método **xticks()** o **yticks()** se cambian los valores a la hora de visualizar el gráfico.

```python
gastos = [50, 120, 80, 95]
meses = ['Enero', 'Febrero', 'Marzo', 'Abril']
mapeo = range(len(meses))

plt.plot(gastos)
plt.xticks(mapeo, meses)  # Mapeado de valores por posicion 0=Enero, 1=Febrero, 2=Marzo, 3=Abril

print(list(mapeo))
```

## Eje X y eje Y
Hasta ahora los gráficos que hemos visualizado los hemos hecho a partir de una sola lista. Al utilizar un parámetro en el método **.plot()** éste lo toma como los valores del eje Y, dibujando el eje X a partir de la longitud de la lista que pasamos como parámetro.

Sin embargo nosotros podemos decidir los valores del eje X si los pasamos en un primer parámetro:

```python
x = [0, 5, 10, 15]
y = gastos

plt.plot(x, y)
```

## Gráfico invertido
Desde el momento que sabemos que estamos enviando los valores de los dos ejes por orden (X y luego Y), podemos cambiar el orden y como resultado obtener un gráfico con los ejes invertidos.

```python
plt.plot(y, x)
```
