```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Gráfico lineal (plot)
Es con el que hemos estado trabajando hasta ahora. No requieren indicar nada específico a la hora de crearlos.

Documentación: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.plot.html

```python
from random import sample

x = range(10)
y = sample(x, len(x))

plt.plot(x, y)
```

## Gráfico de líneas verticales (stem)
Este gráfico es muy útil para ilustrat valores positivos y negativos.

Documentación: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.stem.html

```python
from random import sample

x = range(-10, 11)
y = sample(x, len(x))

plt.stem(x, y)
```

## Gráfico de series comparadas (fill_between)
Con esta opción podemos rellenar el espacio que hay entre dos series de Y.

Documentación: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.fill_between.html

```python
x = range(1,11)
y0 = [0 for n in x]  # Esta serie nos servirá de base para rellenar
y1 = [n**1 for n in x]
y2 = [n**1.5 for n in x]  
y3 = [n**2 for n in x]

plt.fill_between(x, y0, y1, color='orange', interpolate=True)
plt.fill_between(x, y1, y2, color='cyan', interpolate=True)
plt.fill_between(x, y2, y3, color='green', interpolate=True)
```

## Gráfico circular o de pastel (pie)
Este tipo de gráfico es muy útil para representar porcentajes.

Documentación: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.pie.html

```python
# https://es.wikipedia.org/wiki/Anexo:Destinos_tur%C3%ADsticos_mundiales

turistas = [83, 80, 75.3, 57, 55.2, 37, 36.5, 35.2, 35, 33.4]
paises = ['Francia', 'EEUU', 'España', 'China', 'Italia', 'Turquía', 'Alemania', 'México', 'Reino Unido', 'Rusia']

explode = [0,0,0.2,0,0,0,0,0.4,0,0]  # Destacar algunos

plt.pie(turistas, labels=paises, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('TOP 10 DESTINOS MUNDIALES')
plt.xlabel('De un total de {} millones de turistas'.format(sum(turistas)))
```

## Gráfico de cajas y bigotes (boxplot)
Este tipo de gráfico nos permite hacernos una idea de los valores medios en gracias a las cajas, mientras por otro lado los bigotes nos indican los valores límitantes.

Documentación: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
from random import uniform

# https://es.wikipedia.org/wiki/Estatura

jap = [uniform(165,175) for n in  range(100)]
ale = [uniform(175,185) for n in  range(100)]
arg = [uniform(170,180) for n in  range(100)]

plt.boxplot([jap, ale, arg])
plt.xticks([1,2,3], ['Japón', 'Alemania', 'Argentina'])
plt.ylabel('Estatura (cm)')
```
