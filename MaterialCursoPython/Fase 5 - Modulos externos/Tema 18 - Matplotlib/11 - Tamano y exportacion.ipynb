# Tamaño
Hay dos opciones básicas que nos servirán para controlar el tamaño de una figura.
* figsize
* dpi

Estos parámetros se definen justo al momento de crear la figura.

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from random import sample

# figsize=(4,2): tupla con el tamaño de la figura en pulgadas de ancho y alto
# dpi=150: densidad de píxeles por pulgada a la hora, cuantos más dpi más calidad y también mayor tamaño
fig = plt.figure(figsize=(4,2), dpi=150)  
                       
rect = (0, 0, 1, 1)
axes = fig.add_axes(rect)
axes.set_title("Mi gráfico")

# Datos de ejemplo
x = range(1,11)
y1 = sample(range(20), len(x))
y2 = sample(range(20), len(x))

axes.plot(x, y1, 'g--')
axes.plot(x, y2, 'r--')
```

Otro ejemplo haciendo uso de **subplots**:

```python
from random import sample

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6,3), dpi=150)  # Probar también nrows=2, ncols=1, figsize=(3,6)

# Datos de ejemplo
x = range(1,11)
y1 = sample(range(20), len(x))
y2 = sample(range(20), len(x))

axes[0].plot(x, y1, 'g--')
axes[0].set_title("Gráfico 1")

axes[1].plot(x, y2, 'r--')
axes[1].set_title("Gráfico 2")
```

# Exportación
Entendemos por exportación el proceso de generar una imagen a partir de un gráfico para guardarla en el disco duro.

Este proceso se puede hacer relativamente fácil en Jupyter haciendo clic derecho sobre un gráfico y poniendo guardar como, pero también podemos automatizarlo, algo clave para poder visualizar gráficos en tiempo real dentro de aplicaciones y páginas web.

Exportar un gráfico es tan sencillo como llamar al método **savefig** de la *figura* o del propio módulo *plt*.

```python
# Por defecto se guarda en formato PNG
fig.savefig("11 - Mis graficos", dpi=100)

# Con el módulo Pillow instalado para el manejo de imágenes (en Jupyter viene incluido) podemos guardarlo en JPG
# No ofrece tanta calidad pero ocupa menos espacio, sobretodo en gráficos con muchos datos
fig.savefig("11 - Mis graficos.jpg", dpi=100)
```
