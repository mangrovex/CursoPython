# Primer gráfico
Empezaremos nuestra andadura por el más simple, el de líneas. Primero vamos a importar matplotlib, y luego utilizaremos la cláusula **%matplotlib inline** para permitir a Jupyter cargar gráficos en nuestro bloc de notas.

```python
import matplotlib.pyplot as plt
```

Ahora vamos a suponer que queremos crear un gráfico sobre los gastos que tenemos durante por ejemplo 3 meses. Así que vamos a empezar declarando una lista con estos gastos:

```python
gastos = [50, 120, 80, 95]
```

Ahora vamos a utilizar matplotlib para que dibuje un gráfico a partir de estos gastos:

```python
plt.plot(gastos)
```

En este punto tenemos el gráfico cargado en la memoria, sólo nos resta indicarlo a matplotlib que lo muestre:

```python
plt.show()
```

¿Qué os parece? ¿Os esperábais algo mucho más complejo? 

Como véis el gráfico tiene dos ejes, el horizontal (X) con una escala de números y el vertical con los gastos (Y). Estos números corresponden a los índices de los valores en la lista y matplotlib genera una escala a partir de ellos.

El primer elemento con índice 0 es el 50, el segundo de índice 1 el 120, y así hasta el tercero.

## Uso de %matplotlib inline
En Jupyter Notebook existe unos accesos que nos permiten agrupar todo el código de una celda y procesarlo como si fuera un gráfico único, haciendo que podamos ahorrarnos el **plt.show()** del final. Para hacerlo debemos importar la cláusula **%matplotlib inline**. Lo anterior se podría resumir en una celda:

```python
import matplotlib.pyplot as plt
%matplotlib inline

gastos = [50, 120, 80, 95]

plt.plot(gastos)
```

Es más práctico, pero no olvidéis que tenéis que configurar el gráfico en la misma celda y que esto solo funciona en Jupyter Notebook.
