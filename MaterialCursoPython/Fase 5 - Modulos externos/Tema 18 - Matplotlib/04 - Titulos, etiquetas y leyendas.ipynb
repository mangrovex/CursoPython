# Títulos, etiquetas y leyendas

Esta tercera lección la vamos a dedicar a la personalización de nuestros gráficos. Aprendremos a:

* A añadir títulos y etiquetas
* A mostrar leyendas de series

```python
import matplotlib.pyplot as plt
%matplotlib inline  
```

## Títulos y etiquetas
Aunque un gráfico esté muy bien hecho si no mostramos de alguna forma de qué trata no nos servirá de nada. 

Los elementos más básicos para mostrar información son el título y por otro las etiquetas, que sirven para representar qué es cada eje.

```python
x = range(1,11)  # Números del 1 al 10
y = [n**2 for n in x]  # Potencias de 2 del 1 al 10

plt.plot(x, y)

plt.title('Potencias de 2')

plt.xlabel('Número')
plt.ylabel('Resultado')
```

## Leyendas
Otro elemento de información básico es una leyenda. La leyenda añade información sobre las series que estamos representando en el gráfico. Sin embargo para poder mostrar esta información tenemos que darle una **label** (etiqueta) a cada serie.

Posiciones manuales para dibujar una leyenda (0 al 10): https://matplotlib.org/1.3.1/api/pyplot_api.html#matplotlib.pyplot.legend

```python
plt.plot(x, y, label="A la 2")  # Este el momento donde añadimos la label

plt.title('Potencias')
plt.xlabel('Número')
plt.ylabel('Resultado')

plt.legend()  # con el parámetro loc podemos establecer una posición del 0 a 10 (0 es por defecto, 10 es el centro)
```

Como habéis podido notar hemos introducido un nuevo concepto, el de serie.
