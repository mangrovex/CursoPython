# Visualización de múltiples series
Una serie es básicamente un grupo de valores, y podemos representar tantas series como necesitemos en el mismo gráfico.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = range(1,11)
y0 = [n**0 for n in x]
y1 = [n**1 for n in x]
y2 = [n**2 for n in x]

plt.plot(x, y0, label="A la 0")
plt.plot(x, y1, label="A la 1")
plt.plot(x, y2, label="A la 2")

plt.title('Potencias')

plt.xlabel('Número')
plt.ylabel('Resultado')

plt.legend()
```

Como véis Matplotlib añade automáticamente colores distintos a cada línea para que podamos diferenciar las series. Lo bueno es que podemos personalizar estas líneas, cambiando su color y su estilo cuando queremos. Lo veremos en la siguiente lección.
