# Límites
En algunas ocasiones queremos manipular los límites inferiores y superiores de los ejes. Eso podemos hacerlo gracias a los límites:

* xlim
* ylim

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = range(1,11)        # Números del 1 al 10
y = [n**2 for n in x]  # Potencias de 2 del 1 al 10

plt.plot(x, y)
```

Al cambiar los límites las variaciones que percibimos del gráfico cambiarán. 

Si los aumentamos parecerá que nos alejamos haciendo zoom hacia fuera.

```python
plt.plot(x, y)
plt.xlim(0, 20)   # Sólo tenemos 10 valores
plt.ylim(0, 200)  # El máximo es 10**2 que es 100
```

Por otro lado si los reducimos parecerá que nos acercamos, haciendo zoom hacia dentro. 

Esta forma es muy útil para enfocarnos en una sección del gráfico.

```python
plt.plot(x, y)
plt.xlim(len(x)-5, len(x))   # Nos enfocamos en los últimos 5 números
```

Saber jugar con los límites nos ayudará a visualizar los gráficos en la porción que necesitemos en todo momento.
