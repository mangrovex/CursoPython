# Subgráficos en figuras
Hay dos formas básicas de crear subgráficos: 

* Posicionamiento libre 
* Posicionamiento en cuadrícula

## Posicionamiento libre
Este método no crea exactamente subgráficos, si no que se basa en dibujar gráficos directamente sobre el espacio de la figura controlando nosotros mismos su tamaño y posición.

Cada vez que llamamos al método **add_axes** añadimos un nuevo set de ejes a la figura.

Esta forma es muy útil si necesitamos dibujar gráficos dentro de otros gráficos.

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
fig = plt.figure()

x = range(1,11)
y1 = sample(range(20), len(x))
y2 = sample(range(20), len(x))

# Primer gráfico
rect1 = (0, 0, 1, 1)
axes1 = fig.add_axes(rect1)
axes1.plot(x, y1)
axes1.set_title("Gráfico grande")

# Segundo gráfico
rect2 = (0.1, 0.15, 0.3, 0.3)
axes2 = fig.add_axes(rect2)
axes2.set_title("Gráfico pequeño")
axes2.plot(x, y2, 'r--')
```

## Poscionamiento en cuadrícula
La alternativa al posicionamiento libre es el posicionamiento en cuadrícula. 

Este método se basa en crear una cuadrícula utilizando el método **subplots**, que nos devolverá de golpe la figura y los ejes, y además podemos pasarle dos argumentos clave: **nrows** y **ncols**, haciendo referencia al número de filas y columnas de esta cuadrícula.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
        
x = range(1,11)
y1 = sample(range(20), len(x))
y2 = sample(range(20), len(x))
y3 = sample(range(20), len(x))
y4 = sample(range(20), len(x))

axes[0,0].plot(x, y1)
axes[0,0].set_title("Gráfico 1")

axes[0,1].plot(x, y2)
axes[0,1].set_title("Gráfico 2")

axes[1,0].plot(x, y3)
axes[1,0].set_title("Gráfico 3")

axes[1,1].plot(x, y4)
axes[1,1].set_title("Gráfico 4")

# Finalmente ajustamos el contenido para que ocupe el 100% de la figura
fig.tight_layout()
```

Así mismo podemos automatizar este proceso a través de dos bucles for, uno que recorra las filas y otro las columnas para conseguir los ejes exactos donde añadir los datos de cada subgráfico.

```python
fig, axes = plt.subplots(nrows=3, ncols=3)

for row in axes:  # Filas
    for ax in row:  # Columnas
        x = range(1,11)
        y = sample(range(20), len(x))
        ax.plot(x, y)

# Finalmente ajustamos el contenido para que ocupe el 100% de la figura
fig.tight_layout()
```

Evidentemente este ejemplo de automatización no es muy útil, pero si tuviéramos varias fuentes de datos para realizar comparativas podría servirnos para ahorrarnos muchas líneas.
