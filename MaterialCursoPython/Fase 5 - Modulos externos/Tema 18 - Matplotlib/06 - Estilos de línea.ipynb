# Estilos de línea
En esta lección vamos a trabajar como cambiar el color de las líneas, así como su estilo, tamaño y opacidad. 

Podéis encontrar documentacion extendida en la web oficial:
https://matplotlib.org/api/lines_api.html#matplotlib.lines.Line2D.set_linestyle

* **linewidth (lw)**
* **linestyle (ls)**
* **color** (nombre, hexadecimal)
* **alpha** (de 0 a 1)

```python
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
from random import sample

# Vamos a generar 10 muestras aleatorias entre -10 y 10 
x = range(1,11)

y1 = sample(range(-10, 10), len(x))  
y2 = [n+5 for n in y1]  # offset positivo + 5
y3 = [n-5 for n in y1]  # offset negativo - 5

plt.plot(x, y1, ls='-', lw=2, color="#ffb90f", alpha=1, label="n") 
plt.plot(x, y2, 'b--', lw=2, label="n + 5") 
plt.plot(x, y3, 'g-.', lw=2, label="n - 5")  

plt.legend()
```
