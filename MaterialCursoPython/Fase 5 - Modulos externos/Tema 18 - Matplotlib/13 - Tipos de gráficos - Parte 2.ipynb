## Gráfico histograma (hist)
Los histogramas sirven para hacernos una idea de la frecuencia de distribución de valores en un rango determinado. 

Documentación: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.hist.html

```python
from random import uniform

alturas = [uniform(170,180) for n in  range(1000)]

plt.hist(alturas, bins=10, edgecolor='black')  # El rango se define con bins, debe ser menor que el número de muestras

plt.title("Distribución de alturas")
plt.xlabel("Altura media (cm)")
plt.ylabel("Muestras")
```

Otro ejemplo con [distribucion gausiana](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_normal) a partir de un vector normal con Numpy. Este módulo lo veremos en la siguiente unidad pero era interesante ver esta distribución porque tiende a dibujar una campana con su máximo justo en medio.

```python
from numpy.random import normal

numeros = normal(size=1000)

plt.hist(numeros, bins=10, edgecolor='black')

plt.title("Histograma Gaussiano")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
```

## Gráfico de barras (bar)
Este es el clásico gráfico dónde las barras nos indican cantidades en el eje Y para los valores en el eje X.
Documentación: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.bar.html

```python
# http://www.datosmacro.com/prima-riesgo/espana

fechas = ['31/07/2017', '01/08/2017', '02/08/2017', '03/08/2017', '04/08/2017']
primas = [97, 94, 96, 97, 99]

plt.bar(range(5), primas, edgecolor='black')

plt.xticks(range(5), fechas, rotation = 60)
plt.title("PRIMA DE RIESGO EN ESPAÑA")
plt.ylim(90,100)
```

## Gráfico de barras horizontales (barh)
Este es una variante del anterior, es exactamente igual pero tenemos que dar la vuelta a los valores de los ejes para conseguir el mismo comportamiento.

Documentación: https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.barh.html

```python
# http://www.datosmacro.com/prima-riesgo/espana

fechas = ['31/07/2017', '01/08/2017', '02/08/2017', '03/08/2017', '04/08/2017']
primas = [97, 94, 96, 97, 99]

plt.barh(range(5), primas, edgecolor='black')

plt.yticks(range(5), fechas)
plt.title("PRIMA DE RIESGO EN ESPAÑA")
plt.xlim(90,100)
```

## Gráfico de escaleras (step)
Otro gráfico muy simple que nos muestras las variaciones de forma continuada dibujando una escalera. 

https://matplotlib.org/devdocs/api/_as_gen/matplotlib.pyplot.step.html

```python
from random import sample

x = range(10)
y = sample(x, 10)

plt.step(x, y)
```
