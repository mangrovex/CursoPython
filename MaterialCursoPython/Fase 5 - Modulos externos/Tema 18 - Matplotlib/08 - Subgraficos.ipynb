# Subgráficos
Ya hemos visto como crear gráficos e incluso añadirles múltiples series. ¿Pero y si queremos dibujar más de un gráfico en el mismo espacio? ¿Por ejemplo para hacer una comparativa situacional?

En ese caso tenemos a nuestra disposicion los subgráficos.

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

La regla de oro para añadir subgráficos es imaginar que estamos dibujando una tabla. Esta tabla tiene filas y columnas, y cada subgráfico se dibujará en una celda imaginaria dentro de esa tabla.

Lo único que tenemos que tener en cuenta es indicarle a Matplotlib en qué celda debe dibujar el subgráfico.

```python
from random import sample

x = range(1,11)

plt.subplot(1,2,1)  # Tabla 1x2, celda 1
y1 = sample(range(20), 10)
plt.plot(x, y1, 'b')

plt.subplot(1,2,2)  # Tabla 1x2, celda 2
y2 = y1[::-1]  # Lista inversa de y1 con slicing
plt.plot(x, y2, 'g')
```

La misma idea funcionaría para una tabla 2 x 2 con 4 subgráficos. 

Para hacerlo rápido podemos automatizar una generación aleatoria con 2 series en cada uno.

```python
for i in range(4):
    plt.subplot(2,2,i+1)
    y1 = sample(range(20), 10)
    plt.plot(x, y1)
    y2 = sample(range(20), 10)
    plt.plot(x, y2)
```
