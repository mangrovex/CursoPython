# Marcadores
En los gráficos lineales los marcadores hacen referencia a los puntos o vértices donde se dibujan los valores.

Podéis encontrar documentacion extendida en la web oficial: https://matplotlib.org/api/markers_api.html

* **marker**
* **markersize**
* **markerfacecolor**
* **markeredgewidth**
* **markeredgecolor**

````python
import matplotlib.pyplot as plt
%matplotlib inline 
````

````python
from random import sample

x = range(1,11)
y1 = sample(range(50), len(x))
y2 = sample(range(50), len(x))

plt.plot(x, y1, marker='o', markersize=10, markerfacecolor="#ffb90f", markeredgewidth="1", markeredgecolor="black", lw=2)
plt.plot(x, y2, marker='*', markersize=15, markerfacecolor="red", markeredgewidth="1", markeredgecolor="black", lw=2)
````

Si queremos utilizar algun marcador avanzado deberemos importarlos y hacer referencia a ellos de la siguiente manera:

```python
from matplotlib import markers
plt.plot(x, y, marker=markers.TICKLEFT....)
```
